TAREA: Captura una o varias imgenes con monedas no solapadas. Tras visualizar la imagen, identifica de forma interactiva (por ejemplo haciendo clic en la imagen) una moneda de un valor determinado en la imagen (por ejemplo de 1). Tras ello, la tarea se resuelve mostrando por pantalla el nmero de monedas y la cantidad de dinero presentes en la imagen. No hay restricciones sobre utilizar medidas geomtricas o de color. Qu problemas han observado?

Nota: Para establecer la corespondencia entre pxeles y milmetros, comentar que la moneda de un euro tiene un dimetro de 23.25 mm. la de 50 cntimos de 24.35, la de 20 cntimos de 22.25, etc. 

Extras: Considerar que la imagen pueda contener objetos que no son monedas y/o haya solape entre las monedas. Demo en vivo.



```python


# 1. Cargar la imagen de las monedas
img = cv2.imread('Monedas.jpg')
height, width = img.shape[:2]

# Comprobar si la imagen tiene las dimensiones deseadas

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 2. Suavizar la imagen para eliminar altas frecuencias
pimg = cv2.medianBlur(gris, 7)

# 3. Detectar crculos usando HoughCircles
circ = cv2.HoughCircles(
        pimg,  # imagen 
        cv2.HOUGH_GRADIENT,  # tipo de deteccin
        1,
        100,  # distancia mnima entre crculos
        param1=120, # valor del gradiente
        param2=50, # umbral acumulador
        minRadius=50,  # radio mnimo
        maxRadius=150,  # radio mximo
    )

# 4. Dibujar crculos detectados en la imagen
img_cont = np.zeros(img_rgb.shape)
for det in circ[0]:
    x_coor, y_coor, det_radio = det
    cv2.circle(img_rgb, (int(x_coor), int(y_coor)), int(det_radio), (0, 255, 0), 2)
    cv2.circle(img_cont, (int(x_coor), int(y_coor)), int(det_radio), (255, 255, 255), -1)

# 5. Definir la funcin de clic para seleccionar la moneda
moneda_seleccionada = None

def seleccion_moneda(event, x, y, flags, param):
    global moneda_seleccionada
    if event == cv2.EVENT_LBUTTONDOWN:  # Si se hace clic en la imagen
        for det in circ[0]:  # Recorre los crculos detectados
            x_coor, y_coor, det_radio = det
            # Comprobar si el clic est dentro de alguno de los crculos
            if (x - x_coor) ** 2 + (y - y_coor) ** 2 <= det_radio ** 2:
                moneda_seleccionada = (x_coor, y_coor, det_radio)  # Guardar la moneda seleccionada
                break

# Mostrar la imagen con la deteccin de crculos
cv2.imshow('Monedas Detectadas', img)


# 3. Mostrar la imagen en la ventana
cv2.imshow('Monedas Detectadas', img)
cv2.setMouseCallback('Monedas Detectadas', seleccion_moneda)
cv2.waitKey(0)  # Espera a que se haga clic en la ventana
cv2.destroyAllWindows()

if moneda_seleccionada is not None:
    x_seleccionado, y_seleccionado, radio_seleccionado = moneda_seleccionada
    print(f'Moneda seleccionada en ({x_seleccionado}, {y_seleccionado}) con radio {radio_seleccionado}.')

    # 6. Calcular la relacin pxeles a mm basndose en la moneda seleccionada
    # Definir el dimetro de la moneda seleccionada en mm (ejemplo 1)
    diametro_real_mm = 23.25  # Dimetro de 1 en mm
    factor_conversion = diametro_real_mm / (2*radio_seleccionado)

    # 7. Identificar el valor de cada moneda detectada usando la relacin de conversin
    diametros_monedas = {
        "2": 25.75,
        "1": 23.25,
        "50 cntimos": 24.25,
        "20 cntimos": 22.25,
        "10 cntimos": 19.75,
        "5 cntimos": 21.25,
        "2 cntimos": 18.75,
        "1 cntimo": 16.26,
        # Agregar otros valores de monedas
    }

    valores_monedas = {
        "2": 200,
        "1": 100,
        "50 cntimos": 50,
        "20 cntimos": 20,
        "10 cntimos": 10,
        "5 cntimos": 5,
        "2 cntimos": 2,
        "1 cntimo": 1,
        # Agregar otros valores
    }

    monedas_detectadas = []
    total_dinero = 0.0

    for det in circ[0]:
        _, _, radio = det
        diametro_mm = 2 * radio * factor_conversion
        diam=100
        monedita=0
        # Comparar con los dimetros reales de las monedas para determinar el valor
        for moneda, diametro in diametros_monedas.items():
            if abs(diametro - diametro_mm) < diam:
                diam=abs(diametro - diametro_mm)  # Si el tamao se aproxima, es la moneda
                monedita=moneda
        monedas_detectadas.append(monedita)
        total_dinero += valores_monedas[monedita]
        print(diametro_mm)     

    print(f"Monedas detectadas: {monedas_detectadas}")
    print(f"Cantidad total de dinero: {total_dinero/100} ")
else:
    print("No se ha seleccionado ninguna moneda.")
```

Qu es una matriz de confusin?
Se utiliza para mostrar el comportamiento de un clasificador para las distintas clases conocidas, se relacionan las etiquetas de las muestras anotadas frente a las predichas por el clasificador. Se busca una matriz diagonal, pero la perfeccin es infrecuente.

El siguiente ejemplo, muestra el modo de obtener la matriz de confusin para un hipottico problema con cuatro clases, y valores de anotacin (variable y) y prediccin (variable y_pred) obtenidos de forma aleatoria.

```python

import random
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)


# Numero de muestras
n = 100  
nclases = 4

# A falta de clasificador y conjunto de datos, creamos anotaciones y predicciones de forma aleatoria
# Vector aleatorio con etiquetas anotadas
y = [random.randint(0, nclases - 1) for _ in range(n)]
print('Anotaciones ' , y)

# Vector aleatorio con etiquetas predichas por un supuesto clasificador
y_pred = [random.randint(0, nclases - 1) for _ in range(n)]
print('Predicciones ' , y_pred)

print('Cmo de bien encajan anotacin y prediccin?')

#Clculo de mtricas
accuracy = accuracy_score(y, y_pred)
#Para ms de una clase se define la forma de promediar
precision = precision_score(y, y_pred,average='weighted')
recall = recall_score(y, y_pred,average='weighted')
f1score = f1_score(y, y_pred,average='weighted')

print(f"Accuracy (TP/(n))= {accuracy}")
print(f"Precision (TP/(TP+FP)) = {precision}")
print(f"Recall (TP/(TP+FN)) = {recall}")
print(f"F1 Score (2*(precision*recall)/(precision+recall)) = {f1score}")


conf_matrix = confusion_matrix(y, y_pred)
plt.figure(figsize=(8,8))
sns.set(font_scale = 1.75)#tamaos tipografa
sns.set(font_scale = 3.0)

ax = sns.heatmap(
        conf_matrix, # confusion matrix 2D array 
        annot=True, # Muestra nmeros en las celdas
        fmt='d', # valores enteros
        cbar=False, # sin barra de colores
        cmap='flag', # mapa de colores
        #vmax=175 # contraste de color
    )

#Etiquetas matriz de confusin
label_font = {'size':'25'}
ax.set_xlabel("Predicha", labelpad=-0.75, fontdict=label_font)
ax.set_ylabel("Real/Anotado", labelpad=20, fontdict=label_font)
```

TAREA: Las tres imgenes cargadas en la celda inicial, han sido extraidas de las imgenes de mayor tamao presentes en la carpeta. La tarea consiste en extraer caractersticas (geomtricas y/o visuales) e identificar patrones que permitan distinguir las partculas de cada una de las tres clases, evaluando los aciertos y fallos con las imgenes completas considerando las mtricas mostradas y la matriz de confusin. La matriz de confusin, muestra para cada clase el nmero de muestras que se clasifican correctamente de dicha clase, y el nmero de muestras que se clasifican incorrectamente por cada una de las otras dos clases.

En el trabajo [SMACC: A System for Microplastics Automatic Counting and Classification](https://doi.org/10.1109/ACCESS.2020.2970498), las caractersticas geomtricas utilizadas fueron:

- rea en pxeles
- Permetro en pxeles
- Compacidad (relacin entre el cuadrado del permetro y el rea de la partcula)
- Relacin del rea de la partcula con la del contenedor
- Relacin del ancho y el alto del contenedor
- Relacin entre los ejes de la elipse ajustada
- Definido el centroide, relacin entre las distancias menor y mayor al contorno

Si no se quedan satisfechos con la segmentacin obtenida, es el mundo real, tambin en el README comento tcnicas recientes de segmentacin, que podran despertar su curiosidad.

```python

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)
from sklearn.preprocessing import StandardScaler

# Procesar imgenes y extraer caractersticas
def process_image(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    height, width = image.shape[:2]  # Obtener dimensiones de la imagen (contenedor)
    total_area = height * width      # rea del contenedor
    _, thresh = cv2.threshold(image, 90, 255, cv2.THRESH_BINARY_INV)
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    features = []
    for cnt in contours:
        area = cv2.contourArea(cnt)
        perimeter = cv2.arcLength(cnt, True)
        if area == 0:
            continue
        # Calcular las caractersticas geomtricas
        compacity = (perimeter ** 2) / area
        x, y, w, h = cv2.boundingRect(cnt)
        width_height_ratio = w / h if h != 0 else 0
        area_ratio = area / total_area  # Relacin del rea de la partcula con el rea del contenedor

        # Ajustar una elipse si hay suficientes puntos en el contorno
        if len(cnt) >= 5:
            ellipse = cv2.fitEllipse(cnt)
            (center, axes, angle) = ellipse
            major_axis = max(axes)
            minor_axis = min(axes)
            #Mide lo elptico o circular es el objeto basndose en la relacin entre el eje mayor y menor.
            ellipse_ratio = minor_axis / major_axis if major_axis != 0 else 0
        else:
            ellipse_ratio = 0
        
        # Calcular las distancias desde el centroide hasta el contorno
        moments = cv2.moments(cnt)
        if moments["m00"] != 0:
            cx = int(moments["m10"] / moments["m00"])
            cy = int(moments["m01"] / moments["m00"])
        else:
            cx, cy = 0, 0
        #Evala la distribucin de la forma del contorno calculando la relacin entre las distancias ms corta y ms larga desde el centroide hasta el borde del objeto.
        distances = [cv2.pointPolygonTest(cnt, (cx, cy), True) for point in cnt]
        if distances:
            max_distance = max(distances)
            min_distance = min(distances)
            distance_ratio = min_distance / max_distance if max_distance != 0 else 0
        else:
            distance_ratio = 0

        # Aadir todas las caractersticas geomtricas a la lista
        features.append([
            area, perimeter, compacity, width_height_ratio, 
            ellipse_ratio, area_ratio, distance_ratio
        ])
    return features

# Cargar imgenes y extraer caractersticas
pel_features = process_image('PEL.png')
tar_features = process_image('TAR.png')
fra_features = process_image('FRA.png')

# Asignar etiquetas
pel_labels = [0] * len(pel_features)  # Clase 0: PEL
tar_labels = [1] * len(tar_features)  # Clase 1: TAR
fra_labels = [2] * len(fra_features)  # Clase 2: FRA

# Unir caractersticas y etiquetas
X = np.array(pel_features + tar_features + fra_features)
y = np.array(pel_labels + tar_labels + fra_labels)

# Normalizacin de caractersticas
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar el clasificador con pesos balanceados
clf = RandomForestClassifier(n_estimators=200, max_depth=15, random_state=42, class_weight='balanced')

clf.fit(X_train, y_train)

# Predicciones
y_pred = clf.predict(X_test)

# Calcular mtricas
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
f1score = f1_score(y_test, y_pred, average='weighted', zero_division=0)

print(f"Accuracy (TP/(n))= {accuracy}")
print(f"Precision (TP/(TP+FP)) = {precision}")
print(f"Recall (TP/(TP+FN)) = {recall}")
print(f"F1 Score (2*(precision*recall)/(precision+recall)) = {f1score}")

# Matriz de confusin
conf_matrix = confusion_matrix(y_test, y_pred)

# Graficar matriz de confusin
plt.figure(figsize=(8, 8))
sns.set(font_scale=1.75)  # Tamaos tipografa
sns.set(font_scale=3.0)

ax = sns.heatmap(
    conf_matrix,  # Matriz de confusin 2D
    annot=True,   # Muestra nmeros en las celdas
    fmt='d',      # Valores enteros
    cbar=False,   # Sin barra de colores
    cmap='flag'   # Mapa de colores
)

# Etiquetas de la matriz de confusin
label_font = {'size': '25'}
ax.set_xlabel("Predicha", labelpad=-0.75, fontdict=label_font)
ax.set_ylabel("Real/Anotado", labelpad=20, fontdict=label_font)
plt.show()


```
